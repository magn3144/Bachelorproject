```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the csv file
df = pd.read_csv('validation_loss_r_epochs.csv', header=None)

# Create a matrix of validation loss values for all combinations of r and e
r_vals = [1,2,4,8,16,32]
e_vals = [1,2,3,4,5,6]
validation_loss_matrix = np.zeros((len(r_vals), len(e_vals)))
for i in range(len(r_vals)):
    for j in range(len(e_vals)):
        model_name = f"finetuned_r{r_vals[i]}_e{e_vals[j]}"
        validation_loss = float(df.loc[df[0] == model_name][1].values[0])
        validation_loss_matrix[i][j] = validation_loss

# Visualize the matrix
plt.imshow(validation_loss_matrix, cmap='hot', interpolation='nearest')
plt.xlabel('e')
plt.ylabel('r')
plt.title('Validation Loss Matrix')
plt.xticks(np.arange(len(e_vals)), e_vals)
plt.yticks(np.arange(len(r_vals)), r_vals)
plt.colorbar()
#plt.savefig('validation_loss_matrix.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the csv file
df = pd.read_csv('validation_loss_lr_epochs.csv', header=None)

# Create a matrix of validation loss values for all combinations of r and e
lr_values = [0.4 * 6e-4, 0.6 * 6e-4, 0.8 * 6e-4, 6e-4, 1.2 * 6e-4, 1.4 * 6e-4]
lr_values = [round(lr, 7) for lr in lr_values]
epochs_values = [1, 2, 3, 4, 5, 6]
validation_loss_matrix = np.zeros((len(lr_values), len(epochs_values)))
for i in range(len(lr_values)):
    for j in range(len(epochs_values)):
        model_name = f"finetuned_lr{lr_values[i]}_e{epochs_values[j]}"
        validation_loss = float(df.loc[df[0] == model_name][1].values[0])
        validation_loss_matrix[i][j] = validation_loss

# Visualize the matrix
plt.imshow(validation_loss_matrix, cmap='hot', interpolation='nearest')
plt.xlabel('epochs')
plt.ylabel('lr')
plt.title('Validation Loss Matrix')
plt.xticks(np.arange(len(epochs_values)), epochs_values)
plt.yticks(np.arange(len(lr_values)), lr_values)
plt.colorbar()
#plt.savefig('validation_loss_matrix.png')
plt.show()
```
